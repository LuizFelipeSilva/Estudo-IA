```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dict_faturamento = {
    'data_ref': ['2023-01-01', '2020-02-01', '2021-03-01', '2022-04-01', '2023-05-01','2023-06-01', '2020-07-01', '2021-08-01', '2022-09-01', '2023-10-01','2022-11-01', '2023-12-01',],
    'valor': [400000, 890000, 760000, 430000, 920000,340000, 800000, 500000, 200000, 900000,570000, 995000,]
}
```

```python
dict_faturamento
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

## Trazer a média das vendas:


```python
df_faturamento['valor'].mean()
```

```python
df_faturamento['Mês'] = df_faturamento['data_ref'].dt.strftime('%B')
```

## Criar um gráfico de barras vertical mostrando o mês de referência e o valor


```python
df_faturamento.plot.bar(x='data_ref', y='valor')
```

```python
mes_referencia = df_faturamento.loc[df_faturamento['valor'].idxmax()]
```

```python
df_mes_referencia = pd.DataFrame({'Mês':[mes_referencia['Mês']], 'Valor': [mes_referencia['valor']]})
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df_mes_referencia['Mês'], df_mes_referencia['Valor'], width=0.2)  # Reduzindo a largura das barras
plt.xlabel('Mês')
plt.ylabel('Valor')
plt.title(f'Mês com Maior Valor de Faturamento: {df_mes_referencia["Mês"][0]}')
plt.ylim(0, df_mes_referencia['Valor'].values[0] * 1.2)  # Ajustar o limite do eixo y para melhor visualização
plt.xlim(-0.4, 0.4)  # Ajustar os limites do eixo x para aumentar o espaço ao redor da barra
plt.tight_layout()
plt.show()
```

## Criar um gráfico de linhas mostrando o mês de referência e o valor.


```python
df_faturamento = df_faturamento.sort_values(by='data_ref')
```

```python
df_faturamento.plot.line(x='data_ref', y='valor')
```

```python
df = pd.DataFrame(dict_faturamento)
```

```python
indice_max_valor = df['valor'].idxmax()
```

```python
mes_max_valor = df.loc[indice_max_valor, 'data_ref']
```

```python
valor_max = df.loc[indice_max_valor, 'valor']
```

```python
plt.figure(figsize=(6, 4))
plt.hlines(y=valor_max, xmin=min(df['data_ref']), xmax=mes_max_valor, linestyle='-')
plt.scatter(mes_max_valor, valor_max, zorder=5)  # Ponto vermelho para o valor máximo
plt.xlabel('Data de Referência')
plt.ylabel('Valor')
plt.title(f'Valor Máximo de Faturamento: {df_mes_referencia["Mês"][0]}')
plt.grid(True)
plt.tight_layout()
plt.show()
```
